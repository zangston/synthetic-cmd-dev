```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
e_ff_sigma1p0 = [0.03, 0.1, 0.3, 1.0]
spread_sigma1p0 = [0.2121, 0.2098, 0.1997, 0.2071]

e_ff_sigma0p1 = [0.3, 1.0]
spread_sigma0p1 = [0.1962, 0.2058]
```

```python
plt.scatter(e_ff_sigma1p0, spread_sigma1p0, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff_sigma0p1, spread_sigma0p1, color='orange', marker='s', label='sigma = 0.1')
plt.title("Average Spread Across Color Bins, with Binaries")
plt.legend()
```

```python
spread_sigma1p0_nb = [0.1546, 0.0955, 0.0816, 0.0795]
spread_sigma0p1_nb = [0.1220, 0.0867]
```

```python
plt.scatter(e_ff_sigma1p0, spread_sigma1p0_nb, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff_sigma0p1, spread_sigma0p1_nb, color='orange', marker='s', label='sigma = 0.1')
plt.title("Average Spread Across Color Bins, without Binaries")
plt.legend()
```
