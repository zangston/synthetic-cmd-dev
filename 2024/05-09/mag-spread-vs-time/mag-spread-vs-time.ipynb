```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt

from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
```

```python
cluster_times = [0.5, 1, 2, 5, 10, 15, 20]
cluster_times_no_0p5 = [1, 2, 5, 10, 15, 20] # use for sigma 0.1 e_ff 0.01r/u, 0.03r/u
eff_label = ["0.01", "0.03", "0.10", "0.30", "1.00", "Instant"]

# resolved binaries
sigma1p0_resolved = []
sigma1p0_resolved.append([0.05, 0.05, 0.19, 0.27, 0.45, 0.50, 0.45]) # eff001
sigma1p0_resolved.append([0.05, 0.05, 0.16, 0.09, 0.48, 0.42, 0.42]) # eff003
sigma1p0_resolved.append([0.04, 0.05, 0.07, 0.08, 0.45, 0.43, 0.40]) # eff010
sigma1p0_resolved.append([0.06, 0.05, 0.05, 0.08, 0.49, 0.50, 0.50]) # eff030
sigma1p0_resolved.append([0.06, 0.05, 0.04, 0.08, 0.45, 0.42, 0.37]) # eff100
sigma1p0_resolved.append([0.06, 0.05, 0.04, 0.08, 0.45, 0.45, 0.36]) # instant

sigma0p1_resolved = []
sigma0p1_resolved.append([0.05, 0.23, 0.40, 0.53, 0.58, 0.66])
sigma0p1_resolved.append([0.06, 0.20, 0.44, 0.42, 0.38, 0.39])
sigma0p1_resolved.append([0.03, 0.05, 0.19, 0.14, 0.44, 0.31, 0.38])
sigma0p1_resolved.append([0.04, 0.05, 0.11, 0.08, 0.43, 0.31, 0.38])
sigma0p1_resolved.append([0.06, 0.05, 0.05, 0.08, 0.49, 0.51, 0.42])
sigma0p1_resolved.append([0.06, 0.04, 0.04, 0.08, 0.42, 0.40, 0.39])
```

```python
# unresolved binaries

sigma1p0_unresolved = []
sigma1p0_unresolved.append([0.31, 0.27, 0.42, 0.41, 0.54, 0.57, 0.55])
sigma1p0_unresolved.append([0.27, 0.41, 0.48, 0.35, 0.56, 0.53, 0.53])
sigma1p0_unresolved.append([0.09, 0.43, 0.43, 0.35, 0.53, 0.54, 0.48])
sigma1p0_unresolved.append([0.26, 0.47, 0.42, 0.31, 0.58, 0.58, 0.58])
sigma1p0_unresolved.append([0.23, 0.48, 0.43, 0.36, 0.53, 0.53, 0.48])
sigma1p0_unresolved.append([0.36, 0.55, 0.37, 0.29, 0.52, 0.50, 0.49])

sigma0p1_unresolved = []
sigma0p1_unresolved.append([0.24, 0.46, 0.46, 0.64, 0.65, 0.73])
sigma0p1_unresolved.append([0.46, 0.53, 0.59, 0.50, 0.49, 0.48])
sigma0p1_unresolved.append([0.14, 0.42, 0.53, 0.34, 0.56, 0.44, 0.48])
sigma0p1_unresolved.append([0.22, 0.44, 0.47, 0.33, 0.54, 0.45, 0.47])
sigma0p1_unresolved.append([0.30, 0.46, 0.44, 0.32, 0.59, 0.59, 0.53])
sigma0p1_unresolved.append([0.43, 0.57, 0.41, 0.34, 0.50, 0.49, 0.46])
```

```python
cmap = plt.get_cmap('Blues')

for i in range(len(sigma1p0_resolved)):
    color = cmap(i / (len(sigma1p0_resolved) - 1))
    
    if i == 5:
        plt.plot(cluster_times, sigma1p0_resolved[i], color='green', 
                 label='$\Sigma_{cloud} = 1.0 g/cm^2$, $\epsilon_{ff} = \infty$')
    else:
        plt.plot(cluster_times, sigma1p0_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 1.0 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])

cmap = plt.get_cmap('Oranges')

for i in range(len(sigma0p1_resolved)):
    color = cmap(i / (len(sigma0p1_resolved) - 1))
    
    if i == 5:
        plt.plot(cluster_times, sigma0p1_resolved[i], color='green', 
                 label='$\Sigma_{cloud} = 0.1 g/cm^2$, $\epsilon_{ff} = \infty$')
    elif i == 0 or i == 1:
        plt.plot(cluster_times_no_0p5, sigma0p1_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 0.1 g/cm^2$, $\epsilon_{ff} = \infty$' + eff_label[i])
    else:
        plt.plot(cluster_times, sigma0p1_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 0.1 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])


plt.xlabel('Time, T (Myr)')
plt.ylabel('Average magnitude spread, $\Delta m$')
plt.title("Average Spread versus Cluster Time, binaries resolved")
plt.legend(bbox_to_anchor=(1.1, 1.05), ncol=2)

plt.show()
```
