```python
# Get age spread numbers

from nbody6tools import Reader
from nbody62spisea import converter

age_sigma0p1_resolved = []

eff_names = ['001', '003', '010', '030', '100']
cluster_times = [0.5, 1, 2, 5, 10, 15, 20]

for e in range(len(eff_names)):
    path_to_data = '../../../../../../../storage/jfarias/Data/Orion/newbinaries/M3000sigma0.1/sfeff' + eff_names[e] + '_00/'
    
    age_spread_vs_time_arr = []
    
    for t in range(len(cluster_times)):
        snapshot = Reader.read_snapshot(path_to_data, time = cluster_times[t])
        snapshot.to_physical()
        cluster_table = converter.to_spicea_table(snapshot)
        ages = cluster_table['age']
        
        age_spread = np.percentile(ages, 95) - np.percentile(ages, 5)
        age_spread_vs_time_arr.append(age_spread)
    
    age_sigma0p1_resolved.append(age_spread_vs_time_arr)
    
for i in range(len(age_sigma0p1_resolved)):
    plt.plot(cluster_times, age_sigma0p1_resolved[i], label='$\epsilon_{ff}$ = ' + str(eff_names[i]))

plt.xlabel('Time, T (Myr)')
plt.ylabel('Age spread, $\Delta T$ (Myr)')
plt.title('Age spread vs. Cluster Time for $\Sigma_{cloud}$ = 0.1 g/cm$^2$')
plt.legend()
```
