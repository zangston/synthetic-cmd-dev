```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
e_ff = [0.01, 0.03, 0.1, 0.3, 1.0]
e_ff_no_0p01 = [0.03, 0.1, 0.3, 1.0] # use for sigma 0.1ub

# 2 Myr model
spread_sigma0p1_rb = [0.23, 0.20, 0.19, 0.11, 0.05]
spread_sigma0p1_rb_instant = 0.04
spread_sigma1p0_rb = [0.19, 0.16, 0.07, 0.05, 0.04]
spread_sigma1p0_rb_instant = 0.04

spread_sigma0p1_ub = [0.46, 0.53, 0.53, 0.47, 0.44]
spread_sigma0p1_ub_instant = 0.41
spread_sigma1p0_ub = [0.42, 0.48, 0.43, 0.42, 0.43]
spread_sigma1p0_ub_instant = 0.37
```

```python
plt.plot(e_ff, spread_sigma1p0_rb, color='blue', marker='o', label='$\Sigma_{cloud}$ = 1.0 g/cm$^2$, binaries resolved')
plt.plot(e_ff, spread_sigma0p1_rb, color='orange', marker='s', label='$\Sigma_{cloud}$ = 0.1 g/cm$^2$, binaries resolved')
plt.plot(e_ff, spread_sigma1p0_ub, color='teal', marker='*', label='$\Sigma_{cloud}$ = 1.0 g/cm$^2$, binaries unresolved')
plt.plot(e_ff, spread_sigma0p1_ub, color='red', marker='^', label='$\Sigma_{cloud}$ = 0.1 g/cm$^2$. binaries unresolved')
plt.axhline(y=spread_sigma1p0_rb_instant, color='purple', ls='--', label='Spread for $\epsilon_{ff} = \infty$, binaries resolved')
plt.axhline(y=spread_sigma1p0_ub_instant, color='green', ls='--', label='Spread for $\epsilon_{ff} = \infty$, binaries unresolved')

plt.xlabel('Star formation efficiency per freefall time, $\epsilon_{ff}$')
plt.xscale('log')

plt.ylim(0, 0.6)
plt.ylabel('Average magnitude spread, $\Delta m$')

plt.title("Average Spread Across Color Bins, Time = 2 Myr")
plt.legend(bbox_to_anchor=(1.05, 1.0))
```
