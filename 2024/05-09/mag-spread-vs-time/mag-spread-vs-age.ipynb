```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt

from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable
```

```python
cluster_times = [0.5, 1, 2, 5, 10, 15, 20]
cluster_times_no_0p5 = [1, 2, 5, 10, 15, 20] # use for sigma 0.1 e_ff 0.01, 0.03; sigma 1.0 e_ff 0.03
eff_label = ["0.01", "0.03", "0.10", "0.30", "1.00"]

# resolved binaries
sigma1p0_resolved = []
sigma1p0_resolved.append([0.09, 0.10, 0.22, 0.21, 0.57, 0.53, 0.44]) # eff001
sigma1p0_resolved.append([0.09, 0.17, 0.14, 0.53, 0.48, 0.47]) # eff003
sigma1p0_resolved.append([0.08, 0.08, 0.08, 0.16, 0.52, 0.48, 0.42]) # eff010
sigma1p0_resolved.append([0.10, 0.08, 0.06, 0.20, 0.56, 0.56, 0.56]) # eff030
sigma1p0_resolved.append([0.10, 0.07, 0.07, 0.19, 0.53, 0.46, 0.45]) # eff100

sigma0p1_resolved = []
sigma0p1_resolved.append([0.09, 0.22, 0.37, 0.54, 0.57, 0.60])
sigma0p1_resolved.append([0.10, 0.24, 0.39, 0.41, 0.41, 0.42])
sigma0p1_resolved.append([0.07, 0.09, 0.22, 0.13, 0.45, 0.38, 0.39])
sigma0p1_resolved.append([0.08, 0.09, 0.11, 0.11, 0.43, 0.40, 0.39])
sigma0p1_resolved.append([0.10, 0.08, 0.07, 0.17, 0.55, 0.51, 0.43])
```

```python
cmap = plt.get_cmap('Blues')

for i in range(len(sigma1p0_resolved)):
    color = cmap(i / (len(sigma1p0_resolved) - 1))
    
    if i == 1:
        plt.plot(cluster_times_no_0p5, sigma1p0_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 1.0 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])
    else:
        plt.plot(cluster_times, sigma1p0_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 1.0 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])

cmap = plt.get_cmap('Oranges')

for i in range(len(sigma0p1_resolved)):
    color = cmap(i / (len(sigma0p1_resolved) - 1))
    
    if i == 0 or i == 1:
        plt.plot(cluster_times_no_0p5, sigma0p1_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 0.1 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])
    else:
        plt.plot(cluster_times, sigma0p1_resolved[i], color=color, 
                 label='$\Sigma_{cloud} = 0.1 g/cm^2$, $\epsilon_{ff}$ = ' + eff_label[i])


plt.xlabel('Time, T (Myr)')
plt.ylabel('Average magnitude spread, $\Delta m$')
plt.title("Average Spread Across Color Bins, binaries resolved")
plt.legend(bbox_to_anchor=(1.1, 1.05))

plt.show()
```
