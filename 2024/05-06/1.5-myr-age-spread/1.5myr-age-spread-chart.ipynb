```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
e_ff = [0.01, 0.03, 0.1, 0.3, 1.0]
spread_sigma0p1_rb = [0.2036, 0.1519, 0.1637, 0.1217, 0.0768]
spread_sigma1p0_rb = [0.1558, 0.1591, 0.0867, 0.0703, 0.0686]
```

```python
plt.scatter(e_ff, spread_sigma1p0_rb, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff, spread_sigma0p1_rb, color='orange', marker='s', label='sigma = 0.1')
plt.xlabel('Log-e_ff')
plt.xscale('log')
plt.ylabel('Average magnitude spread')
plt.title("Average Spread Across Color Bins, with Resolved Binaries")
plt.legend()
```

```python
spread_sigma0p1_ub = [0.2604, 0.2346, 0.2343, 0.2121, 0.2184]
spread_sigma1p0_ub = [0.2252, 0.2279, 0.2242, 0.2111, 0.2190]
```

```python
plt.scatter(e_ff, spread_sigma1p0_ub, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff, spread_sigma0p1_ub, color='orange', marker='s', label='sigma = 0.1')
plt.xlabel('Log-e_ff')
plt.xscale('log')
plt.ylabel('Average magnitude spread')
plt.title("Magnitude Spread Across Color Bins, with Unresolved Binaries")
plt.legend()
```

```python
plt.scatter(e_ff, spread_sigma1p0_rb, color='blue', marker='o', label='sigma = 1.0, binaries resolved')
plt.scatter(e_ff, spread_sigma0p1_rb, color='orange', marker='s', label='sigma = 0.1, binaries resolved')
plt.scatter(e_ff, spread_sigma1p0_ub, color='teal', marker='*', label='sigma = 1.0, binaries unresolved')
plt.scatter(e_ff, spread_sigma0p1_ub, color='red', marker='^', label='sigma = 0.1. binaries unresolved')
plt.xlabel('Log-e_ff')
plt.xscale('log')
plt.ylabel('Average magnitude spread')
plt.title("Average Spread Across Color Bins")
plt.legend(bbox_to_anchor=(1.05, 1.0))
```
