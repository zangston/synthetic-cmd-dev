```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
```

```python
e_ff = [0.01, 0.03, 0.1, 0.3, 1.0]
spread_sigma0p1_rb = [0.2122, 0.2153, 0.1118, 0.0987, 0.0977]
spread_sigma1p0_rb = [0.1584, 0.0979, 0.1007, 0.0983, 0.1035]
```

```python
plt.scatter(e_ff, spread_sigma1p0_rb, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff, spread_sigma0p1_rb, color='orange', marker='s', label='sigma = 0.1')
plt.xlabel('Log-e_ff')
plt.xscale('log')
plt.ylabel('Average magnitude spread')
plt.title("Average Spread Across Color Bins, with Resolved Binaries")
plt.legend()
```

```python
spread_sigma0p1_ub = [0.2177, 0.2227, 0.1639, 0.1634, 0.1675]
spread_sigma1p0_ub = [0.1825, 0.1768, 0.1801, 0.1626, 0.1774]
```

```python
plt.scatter(e_ff, spread_sigma1p0_ub, color='blue', marker='o', label='sigma = 1.0')
plt.scatter(e_ff, spread_sigma0p1_ub, color='orange', marker='s', label='sigma = 0.1')
plt.xlabel('Log-e_ff')
plt.xscale('log')
plt.ylabel('Average magnitude spread')
plt.title("Magnitude Spread Across Color Bins, with Unresolved Binaries")
plt.legend()
```

```python
plt.scatter(e_ff, spread_sigma1p0_rb, color='blue', marker='o', label='$\Sigma_{cloud}$ = 1.0 g/cm$^2$, binaries resolved')
plt.scatter(e_ff, spread_sigma0p1_rb, color='orange', marker='s', label='$\Sigma_{cloud}$ = 0.1 g/cm$^2$, binaries resolved')
plt.scatter(e_ff, spread_sigma1p0_ub, color='teal', marker='*', label='$\Sigma_{cloud}$ = 1.0 g/cm$^2$, binaries unresolved')
plt.scatter(e_ff, spread_sigma0p1_ub, color='red', marker='^', label='$\Sigma_{cloud}$ = 0.1 g/cm$^2$. binaries unresolved')
plt.xlabel('Log-freefall timescale, $\epsilon_{ff}$')
plt.xscale('log')
plt.ylabel('Average magnitude spread, $\Delta m$')
plt.title("Average Spread Across Color Bins")
plt.legend(bbox_to_anchor=(1.05, 1.0))
```
