```python
# import packages
import numpy as np
import pylab as py
import pdb
import matplotlib.pyplot as plt
import math

# import spisea packages
from spisea import synthetic, evolution, atmospheres, reddening, ifmr
from spisea.imf import imf, multiplicity

# import nbody6tools reader to read data
from nbody6tools import Reader

# import nbody62spisea to retrieve star ages
from nbody62spisea import converter
```

```python
# define isochrone parameters
AKs = 0
dist = 410
atm_func = atmospheres.get_merged_atmosphere
red_law = reddening.RedLawHosek18b()
filt_list = ['jwst,F115W', 'jwst,F182M']
filters = ['m_jwst_F115W', 'm_jwst_F182M']
iso_dir = 'isochrones/'
metallicity = 0

model_iso = []

baraffe = evolution.Baraffe15()
model_iso.append(baraffe)

pisa = evolution.Pisa()
model_iso.append(pisa)

ekstrom = evolution.Ekstrom12()
model_iso.append(ekstrom)

parsec = evolution.Parsec()
model_iso.append(parsec)

merged_model = evolution.MergedBaraffePisaEkstromParsec()
model_iso.append(merged_model)
```

```python
iso_arr = []

age = np.log10(1.5e6)

for i in range(5):
    iso = synthetic.IsochronePhot(age, AKs, dist, metallicity=metallicity,
                            evo_model=model_iso[i], atm_func=atm_func,
                            red_law=red_law, filters=filt_list,
                                iso_dir=iso_dir)
    iso_arr.append(iso)
```

```python
# Plot CMD
fig, axes = py.subplots(figsize=(15, 10))
py.subplot(1, 2, 1)

colors = ['r', 'b', 'g', 'orange', 'k']
labels = ['Baraffe', 'Pisa', 'Ekstrom', 'Parsec', 'Merged']

for i in range(len(iso_arr)):
    if i == 4:
        lw = 4
        zorder = -10
        alpha = 0.5
    else:
        lw = 1
        zorder = None
        alpha = 1
    py.plot(iso_arr[i].points[filters[0]] - iso_arr[i].points[filters[1]], 
           iso_arr[i].points[filters[1]], color=colors[i], label=labels[i], lw = lw, zorder = zorder, alpha = alpha)

idx = np.where(abs(iso_arr[4].points['mass'] - 0.5) == min(abs(iso_arr[4].points['mass'] - 0.5)))[0].item()
py.plot(iso_arr[4].points[filters[0]][idx] - iso_arr[4].points[filters[1]][idx], 
       iso_arr[4].points[filters[1]][idx], 'b*', ms=15, label='0.5 $M_\odot$')

idx = np.where(abs(iso_arr[4].points['mass'] - 6) == min(abs(iso_arr[4].points['mass'] - 6)))[0].item()
py.plot(iso_arr[4].points[filters[0]][idx] - iso_arr[4].points[filters[1]][idx], 
       iso_arr[4].points[filters[1]][idx], 'r*', ms=15, label='6 $M_\odot$')

py.gca().invert_yaxis()
py.legend()
```

```python
mist = synthetic.IsochronePhot(age, AKs, dist, metallicity=metallicity,
                            evo_model=evolution.MISTv1(), atm_func=atm_func,
                            red_law=red_law, filters=filt_list,
                                iso_dir=iso_dir)
```

```python
fig, axes = py.subplots(figsize=(15, 10))
py.subplot(1, 2, 1)

py.plot(mist.points[filters[0]] - mist.points[filters[1]], 
           mist.points[filters[1]], color='r', label='MIST')

py.plot(iso_arr[0].points[filters[0]] - iso_arr[0].points[filters[1]], 
           iso_arr[0].points[filters[1]], color='b', label='Baraffe')

idx = np.where(abs(mist.points['mass'] - 0.5) == min(abs(mist.points['mass'] - 0.5)))[0].item()
py.plot(mist.points[filters[0]][idx] - mist.points[filters[1]][idx], 
       mist.points[filters[1]][idx], 'b*', ms=15, label='0.5 $M_\odot$')


py.gca().invert_yaxis()
py.legend()
```
