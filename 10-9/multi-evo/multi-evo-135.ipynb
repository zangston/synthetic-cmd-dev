```python
# import packages
from spisea import synthetic, evolution, atmospheres, reddening, ifmr
from spisea.imf import imf, multiplicity
import numpy as np
import pylab as py
import pdb
import matplotlib.pyplot as plt
```

```python
# Create isochrone
age_arr = [np.log10(1*10**6.), np.log10(3*10**6.), np.log10(5*10**6.)]
AKs = 3 # extinction in mags
dist = 4500 # distance in parsec
metallicity = 0 # Metallicity in [M/H]

# Evolution/atmosphere models and extinction law
evo_models = [evolution.Baraffe15(), evolution.Ekstrom12(), evolution.Parsec(), evolution.Pisa()]
atm_func = atmospheres.get_merged_atmosphere
red_law = reddening.RedLawHosek18b()

# Specify filters for synthetic photometry
filt_list = ['jwst,F162M', 'jwst,F200W']

# Specify the directory we want the output isochrone
# table saved in. If the directory does not already exist,
# SPISEA will create it.
iso_dir = 'isochrones/'

# Make Isochrone Grid
instances = np.empty((len(evo_models), len(age_arr)), dtype=object)

for i in range(len(evo_models)):
    for j in range(len(age_arr)):
        print("create isochrone with evolution model " + str(i + 1) + " of "  + str(len(evo_models)) + 
              " and age " + str(j + 1) + " of " + str(len(age_arr)))
        my_iso = synthetic.IsochronePhot(age_arr[j], AKs, dist, metallicity=0,
                            evo_model=evo_models[i], atm_func=atm_func,
                            red_law=red_law, filters=filt_list,
                                iso_dir=iso_dir)
        instances[i][j] = my_iso
        
print("isochrone generation done")
```

```python
subplot_r = 4
subplot_c = 2
evo_model_names = ["Baraffe15", "Ekstrom12", "Parsec", "Pisa"]
fig, axes = py.subplots(subplot_r, subplot_c, figsize=(15, 20))
watts_to_lsun = 1.0 / (3.846e26) # conversion factor for watts to Lsun

for i in range(subplot_r):
    for j in range(subplot_c):
        ax = axes[i, j]  # Get the current axis

        # Plot CMD
        if j == 0:
            ax.plot(
                instances[i][0].points["m_jwst_F162M"] - instances[i][0].points["m_jwst_F200W"],
                instances[i][0].points["m_jwst_F200W"],
                "r-",
                label="1 Myr",
            )
            ax.plot(
                instances[i][1].points["m_jwst_F162M"] - instances[i][1].points["m_jwst_F200W"],
                instances[i][1].points["m_jwst_F200W"],
                "b-",
                label="3 Myr",
            )
            ax.plot(
                instances[i][2].points["m_jwst_F162M"] - instances[i][2].points["m_jwst_F200W"],
                instances[i][2].points["m_jwst_F200W"],
                "k-",
                label="5 Myr",
            )
            ax.set_xlabel("F162m - F200W")
            ax.set_ylabel("F200W")
            ax.invert_yaxis()
            ax.legend()
            ax.set_title(evo_model_names[i] + " CMD")

        # Plot HR
        elif j == 1:
            data_iso_Lsun = instances[i][0].points["L"] * watts_to_lsun
            data2_iso_Lsun = instances[i][1].points["L"] * watts_to_lsun
            data3_iso_Lsun = instances[i][2].points["L"] * watts_to_lsun
            ax.plot(
                instances[i][0].points["Teff"], data_iso_Lsun, "r-", label="1 Myr"
            )
            ax.plot(
                instances[i][1].points["Teff"], data2_iso_Lsun, "b-", label="3 Myr"
            )
            ax.plot(
                instances[i][2].points["Teff"], data3_iso_Lsun, "k-", label="5 Myr"
            )
            ax.set_yscale("log")
            ax.set_xlabel("Effective Temperature, $T$ (K)")
            ax.set_ylabel("Luminosity, $L$ ($L_{\odot}$)")
            ax.invert_xaxis()
            ax.legend()
            ax.set_title(evo_model_names[i] + " HR")
py.tight_layout()
```
