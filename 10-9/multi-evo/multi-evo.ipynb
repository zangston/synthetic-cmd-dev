```python
# import packages
from spisea import synthetic, evolution, atmospheres, reddening, ifmr
from spisea.imf import imf, multiplicity
import numpy as np
import pylab as py
import pdb
import matplotlib.pyplot as plt
```

```python
# Create isochrone
# logAge = np.log10(1*10**6.) # Age in log(years)
age_arr = [np.log10(1*10**6.), np.log10(3*10**6.)]
AKs = 3 # extinction in mags
dist = 4500 # distance in parsec
metallicity = 0 # Metallicity in [M/H]

# Evolution/atmosphere models and extinction law
evo_models = [evolution.Baraffe15(), evolution.Ekstrom12(), evolution.Parsec(), evolution.Pisa()]
atm_func = atmospheres.get_merged_atmosphere
red_law = reddening.RedLawHosek18b()

# Specify filters for synthetic photometry
filt_list = ['jwst,F162M', 'jwst,F200W']

# Specify the directory we want the output isochrone
# table saved in. If the directory does not already exist,
# SPISEA will create it.
iso_dir = 'isochrones/'

# Make Isochrone Grid
instances = np.empty((len(evo_models), len(age_arr)), dtype=object)

'''
my_iso = synthetic.IsochronePhot(logAge, AKs, dist, metallicity=0,
                            evo_model=evo_model, atm_func=atm_func,
                            red_law=red_law, filters=filt_list,
                                iso_dir=iso_dir)
'''

for i in range(len(evo_models)):
    for j in range(len(age_arr)):
        print("create isochrone with evolution model " + str(i) " of " str(len(evo_models)) + 
              " and age " + str(j) + " of " + str(len(age_arr)))
        my_iso = synthetic.IsochronePhot(age_arr[j], AKs, dist, metallicity=0,
                            evo_model=evo_models[i], atm_func=atm_func,
                            red_law=red_law, filters=filt_list,
                                iso_dir=iso_dir)
        instances[i][j] = my_iso
        
print("isochrone generation done")
```

```python
# Plot Baraffe CMD
fig, axes = py.subplots(figsize=(15, 10))
py.subplot(1, 2, 1)
#py.figure(1, figsize=(10,10))
#py.clf()
py.plot(instances[0][0].points['m_jwst_F162M'] - instances[0][0].points['m_jwst_F200W'], 
       instances[0][0].points['m_jwst_F200W'], 'r-', label='_nolegend_')
py.plot(instances[0][1].points['m_jwst_F162M'] - instances[0][1].points['m_jwst_F200W'], 
       instances[0][1].points['m_jwst_F200W'], 'b-', label='_nolegend_')
py.xlabel('F162m - F200W')
py.ylabel('F200W')
py.gca().invert_yaxis()
py.legend()

# Plot Baraffe HR
py.subplot(1, 2, 2)
watts_to_lsun = 1.0 / (3.846e26) # conversion factor for watts to Lsun
data_iso_Lsun = instances[0][0].points['L'] * watts_to_lsun
data2_iso_Lsun = instances[0][1].points['L'] * watts_to_lsun

#py.figure(1, figsize=(10,10))
#py.clf()
py.plot(instances[0][0].points['Teff'], data_iso_Lsun, 'r-', label='_nolegend_')
py.plot(instances[0][1].points['Teff'], data2_iso_Lsun, 'b-', label='_nolegend_')
py.yscale('log')
py.xlabel('Effective Temperature, $T$ (K)')
py.ylabel('Luminosity, $L$ ($L_{\odot}$)')
py.gca().invert_xaxis()
py.legend()
```
